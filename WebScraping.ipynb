######  Paginación

```python
import requests
from bs4 import BeautifulSoup
```

```python
def obtener_contenido_pagina(url):
    response=requests.get(url)
    return response.content
```

```python
def analizar_contenido_html(html):
    return BeautifulSoup(html,'html.parser')
```

```python

def procesar_pagina():
    data=[]

    titulos=[]
    precios=[]
    existencias=[]
    ratings=[]

    titulo_items= soup.find_all('h3')

    for items in titulo_items:
        titulo=items.a #accede a la etiqueta a
        titulos.append(titulo['title'])
        
    precio_items= soup.find_all('p',class_="price_color")

    for item in precio_items:
        precio=item.text #se refiere unicamente a la parte de textos
        precios.append(precio)

    existencia_items=soup.find_all('p',class_="instock availability")

    for item in existencia_items:
        existencia=item.text
        existencias.append(existencia)


    rating_items=soup.find_all('p',class_="star-rating")

    for item in rating_items:
        rating=item['class'][1]
        ratings.append(rating)


    data={
        'Título':titulos,
        'Precio':precios,
        'Rating':ratings,
        'Existencia':existencias
    }

    df=pd.DataFrame(data)
```

```python
def manejar_paginacion(url_base,num_paginas):
    for i in range(1,num_paginas+1):
        url=url_base+'/page'+str(i)+'.html'#actualiza la url base con el num de página
        contenido_pagina=obtener_contenido_pagina(url)
        soup=analizar_contenido_html(contenido_pagina)
        procesar_pagina(soup)
```

```python
url_base="http://books.toscrape.com/catalogue/category/books/mystery_3/index.html"
num_paginas=20

manejar_paginacion(url_base,num_paginas)
```

###### Expresiones regulares


```python
html=requests.get('http://books.toscrape.com/')
soup=BeautifulSoup(html.text,'html.parser')
```

```python
import re

item=soup.select('li.current')

patron=r'Page (\d+) of (\d+)'

concuerda= re.search(patron,html.text)

print(concuerda)

if concuerda:
    page_number=int(concuerda.group(1))
    total_pages=int(concuerda.group(2))
    print(page_number,total_pages)
```

```python
url_base="http://books.toscrape.com/index.html"
num_paginas=total_pages

manejar_paginacion(url_base,num_paginas)
```

##### Ejercicio Cyberpuerta

```python
'Peticiones http'
page=requests.get("https://www.cyberpuerta.mx/")
soup= BeautifulSoup(page.text,'html.parser')
print(soup.prettify())
```

```python
items_find=soup.find_all('div',class_='emproduct_cptitleBox')
print(items_find)

titulos_destacados=[]

for item in items_find:
    titulo=item.a #se refiere unicamente a la parte de textos
    titulos_destacados.append(titulo['title'])
```

```python
titulos_destacados
```

```python
precio_found=soup.find_all('label',class_="price")
print(precio_found)

precios=[]
 
for item in precio_found:
    precio=item.text
    precios.append(precio)
```

```python
precios=[item.strip() for item in precios]

precios
```

```python
envio_find=soup.find_all('span',class_='deliveryvalue')
print(envio_find)

envios=[]
for item in envio_find:
    precio_envio=item.text
    envios.append(precio_envio)
    
```

```python
envios
```

```python
piezas_find=soup.find_all('div',class_="emstock")
print(piezas_find)

piezas_dispo=[]

for item in piezas_find:
    piezas=item.span.text
    piezas_dispo.append(piezas)
```

```python
piezas_dispo
```

```python
import pandas as pd
```

```python
data_destacados={
    'Artículo':titulos_destacados,
    'Precio':precios,
    'Costo de envío':envios,
    'Piezas disponibles':piezas_dispo
}

df_destacados=pd.DataFrame(data_destacados)

print(df_destacados)
```
