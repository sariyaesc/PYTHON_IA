```python
import requests
from bs4 import BeautifulSoup
```

###### Enviar petición HTTP para extraer código de fuente de la página

```python
page=requests.get("http://books.toscrape.com/")
soup= BeautifulSoup(page.text,'html.parser')
print(soup.prettify())
```

```python
titulo_items= soup.find_all('h3')

titulos=[]
for items in titulo_items:
    titulo=items.a #accede a la etiqueta a
    titulos.append(titulo['title']) #da con la etiqueta de title directamente
    
print(titulo_items)
print(titulos)
```

```python
titulos
```

```python
precios=[]

precio_items= soup.find_all('p',class_="price_color")

for item in precio_items:
    precio=item.text #se refiere unicamente a la parte de textos
    precios.append(precio)


print(precio_items)
print (precios)
```

```python
precios
```

```python
existencias=[]

existencia_items=soup.find_all('p',class_="instock availability")

for item in existencia_items:
    existencia=item.text
    existencias.append(existencia)

```

```python
existencias
```

```python
ratings=[]
rating_items=soup.find_all('p',class_="star-rating")
print(rating_items)

for item in rating_items:
    rating=item['class'][1]
    ratings.append(rating)

print(ratings)

```

```python
ratings
```

```python
import pandas as pd
```

```python
data={
    'Título':titulos,
    'Precio':precios,
    'Rating':ratings,
    'Existencia':existencias
}

df=pd.DataFrame(data)

print(df)
```
